We can see that the total number of chapters is very low as 5, meaning that an the additional 2 episodes that were added causes the adapt ratio to be much higher. The additional episodes can be completely skipped without problems so we'll remove them from the data. The total episodes for this arc can be edited using python and pandas.

```python
df_arcs.loc['Loguetown Arc', 'TotalEpisodes'] = 4
df_arcs.loc['Loguetown Arc', 'Adapt Ratio'] = 4/5
df_arcs.loc['Loguetown Arc']
```

Now the dataframe is updated to have only 4 total episodes for the lougetown arc. The adapt ratio was also changed to reflect the new data. Now the data can be exported to create a new visualization.
